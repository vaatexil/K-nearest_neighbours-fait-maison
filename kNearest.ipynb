```python
dataset2 =[[1, 2],[1.4, 3], [2, 4.5], [3,5.5], [4, 7.6], [3.5, 7]]
new_feature2 = 2.5

for i in dataset2:
    plt.scatter(i[0], i[1], s= 100, color = "b")
plt.show()
```

```python
def k_nearest_regression(data, predict, k = 2): # k nearest pour classification
    distances = []
    for point in data:
        distances.append(abs(point[0] - predict))
    nearestN = np.argsort(np.array(distances)) # take the nearest neighbours
    sumKN = 0
    for index in nearestN[:k]:
        sumKN += data[index][1]
    return sumKN / k
```

```python
prevision = k_nearest_regression(dataset2, new_feature2)
```
